```python
import pandas as pd
```

```python
name=['허준호', '이가원', '배규민', '고고림', '이새봄', '이보람', '이루리', '오다현']
gender = ['남자', '여자', '남자', '남자', '여자', '여자', '여자', '여자']
age = [30, 24, 23, 21, 28, 26, 24, 24]
height = [183, 162, 179, 182, 160, 163, 157, 172]
data = {'이름':name, '성별':gender, '나이':age, '키':height}
df = pd.DataFrame(data)
df
```

```python
# 1. 몸무게 column 추가 : [ 82.5, 50.0, 78.0 , 80.5, 48.0, 51.2, 46.0, 65.0]
df['몸무게'] = [ 82.5, 50.0, 78.0 , 80.5, 48.0, 51.2, 46.0, 65.0]
df
```

```python
# 2. 키가 180 이상인 사람의 이름과 키를 조회하라.
df.info()
df.loc[df['키'] >= 180, ['이름', '키']]
```

```python
# 3. 여자 중 키가 160보다 큰 사람을 조회하라.
df.loc[(df.성별 == '여자') & (df.키 > 160)]
```

```python
# 4. 성별 키의 평균을 구하라
df.groupby('성별')['키'].mean()
```

```python
# 5. 성별 키와 몸무게의 평균. 합, 인원수를 구하라.
df.groupby('성별').agg({'키':['sum','mean','count'], '몸무게': ['sum','mean','count']})
```
