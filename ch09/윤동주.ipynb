```python
from bs4 import BeautifulSoup as soup
import urllib.request as req
```

```python
url = "https://ko.wikisource.org/wiki/%EC%A0%80%EC%9E%90:%EC%9C%A4%EB%8F%99%EC%A3%BC"
res = req.urlopen(url)
bs = soup(res, 'html.parser')
li_list = bs.select("#mw-content-text > div.mw-content-ltr.mw-parser-output > ul > li")
for li in li_list:
    print(li.a.text)
    li_list = li.select('ul>li')
    if len(li_list) > 0:
        for sli in li_list:
            print('  -', sli.a.text)
```
